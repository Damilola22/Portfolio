I have changed the column names from camelcase to snakecase and assigned them back to the autos dataset, so as to make them more descriptive and readable. 

```python
autos.describe(include="all")
```

The following observations were made:
- There are columns which have one value and need to be dropped.
- The pictures_nr column mostly has one value
- Postal code column
- seller column
- unrepaired damage
- The symbol added to the price column as well as the odometer column needs to be cleaned and stored as float or integer.

```python
#remove irerelevant columns
autos.drop(columns = ["seller", "pictures_nr", "offer_type"], inplace=True)
```

```python
#Removing the non-numeric characters in the price columns
autos["price"] = (autos["price"]
                .str.replace("$", "")
                .str.replace(",", "")
                .astype(float)
                )
#Rename the price column to make it more descriptive
autos.rename({"price": "price_usd"}, axis=1, inplace=True)
```

```python
#Removing the non-numeric charcters in the odometer column
autos["odometer"] = (autos["odometer"]
                   .str.replace(",", "")
                   .str.replace("km", "")
                   .astype(float)
                   )
            
#Rename the odometer column to make it more descriptive
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
```
